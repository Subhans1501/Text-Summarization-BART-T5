```python
import json
import pandas as pd
import os
raw_data_path = "data//raw//arxiv-metadata-oai-snapshot.json"
processed_data_path = "data//processed/arxiv_sample.csv"
sample_size = "all" 
chunk_size = 50000
print(f"Reading massive dataset from {raw_data_path}...")
os.makedirs(os.path.dirname(processed_data_path), exist_ok=True)
if os.path.exists(processed_data_path):
    os.remove(processed_data_path)
    print("Cleaned up old dataset file.")

data = []
total_processed = 0

try:
    with open(raw_data_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if str(sample_size).lower() != "all" and i >= int(sample_size):
                break
                
            paper = json.loads(line)
            data.append({
                "id": paper.get("id"),
                "title": paper.get("title", "").replace("\n", " ").strip(),
                "abstract": paper.get("abstract", "").replace("\n", " ").strip()
            })
            total_processed += 1
            if len(data) >= chunk_size:
                df = pd.DataFrame(data)
                # Only write headers for the very first chunk
                header = not os.path.exists(processed_data_path) 
                df.to_csv(processed_data_path, mode='a', header=header, index=False)
                
                print(f"Processed {total_processed} papers so far...")
                data = [] 
    if len(data) > 0:
        df = pd.DataFrame(data)
        header = not os.path.exists(processed_data_path)
        df.to_csv(processed_data_path, mode='a', header=header, index=False)
        
    print(f"Success! Saved {total_processed} total papers to {processed_data_path}")
    
except FileNotFoundError:
    print(f"Error: Could not find {raw_data_path}. Ensure your 3GB JSON file is there!")
```
